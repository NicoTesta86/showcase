## Trends in terrorism (increases or decreases over time by country or globally)

To analyze trends in terrorism, we can look at the data from both files. The "Overall Scores.csv" provides a high-level view with scores and rankings by country over several years, which can help us identify global trends and changes in the severity of terrorism. The "Combined raw.csv" offers detailed yearly data, including the number of incidents, fatalities, injuries, and hostages, which can be used to observe trends in terrorism's impact over time.

For a comprehensive analysis of trends in terrorism, we can start by:

- Global Trends: Examining the global trend in terrorism scores over the years using data from "Overall Scores.csv". This involves aggregating the scores by year to see if there's a global increase or decrease in terrorism impact.
- Country-Specific Trends: Looking at the change in the number of incidents, fatalities, injuries, and hostages for select countries or top-impacted countries over time. This requires detailed data from "Combined raw.csv".

```python
# Extract year, score columns and calculate the global average score per year
score_columns = [col for col in overall_scores_df.columns if 'Score' in col]
year_columns = [int(col.split(' ')[0]) for col in score_columns]

# Calculate the average score for each year
global_average_scores = overall_scores_df[score_columns].mean()

# Convert year_columns to string for plotting
year_columns_str = [str(year) for year in year_columns]

global_average_scores.index = year_columns_str

global_average_scores
```

The global average terrorism scores from 2012 to 2022 show a trend where the impact of terrorism fluctuated over the years, with a peak around 2017. Starting from an average score of 2.45 in 2012, there was a slight decrease and then an increase, peaking in 2018 at an average score of 2.63. From 2019 onwards, there's a noticeable trend of decrease, reaching the lowest average score of 2.17 by 2022.

This trend indicates that, on average, the global impact of terrorism might have been highest around 2016-2018 and has been decreasing since then.

Next, we can focus on country-specific trends, particularly looking at changes in the number of incidents, fatalities, injuries, and hostages for select countries over time.

To analyze country-specific trends in terrorism, we'll focus on the countries most impacted by terrorism according to the data. We can use the "Combined raw.csv" dataset to identify these countries based on either the total number of incidents or the highest scores over the available years.

We'll proceed by:

1. Identifying the top-impacted countries based on the total number of incidents or scores over the years.
2. Analyzing the trends in the number of incidents, fatalities, injuries, and hostages for these countries over the years.

Let's start by identifying the top-impacted countries based on the total number of incidents from 2012 to the most recent year in the dataset. We'll summarize the data to see which countries have the highest total number of incidents.

```python
# Summarize the data to find the countries with the highest total number of incidents
top_countries_incidents = combined_raw_df.groupby('Country')['Incidents'].sum().sort_values(ascending=False).head(5)

top_countries_incidents
```

Next, we'll analyze the trends in the number of incidents, fatalities, injuries, and hostages for these top-impacted countries over the years. We'll start with Iraq as it has the highest total number of incidents. Let's visualize the yearly trend in the number of incidents, fatalities, injuries, and hostages for Iraq

```python
import matplotlib.pyplot as plt

# Filter data for Iraq
iraq_data = combined_raw_df[combined_raw_df['Country'] == 'Iraq']

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(iraq_data['Year'], iraq_data['Incidents'], marker='o', linestyle='-', label='Incidents')
ax.plot(iraq_data['Year'], iraq_data['Fatalities'], marker='s', linestyle='-', label='Fatalities')
ax.plot(iraq_data['Year'], iraq_data['Injuries'], marker='^', linestyle='-', label='Injuries')
ax.plot(iraq_data['Year'], iraq_data['Hostages'], marker='x', linestyle='-', label='Hostages')

ax.set_title('Terrorism Trends in Iraq (2012-2022)')
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.legend()
plt.xticks(iraq_data['Year'], rotation=45)
plt.grid(True)

plt.show()
```

The graph illustrates the trends in terrorism-related incidents, fatalities, injuries, and hostages in Iraq from 2012 to 2022. We observe several key points from this visualization:

- The number of incidents peaked around the mid-2010s and has since shown a general decline.
- Fatalities and injuries follow a similar trend to incidents, indicating that as the number of incidents decreased, so did the casualties.
- The number of hostages fluctuates year by year but does not show a clear trend like the other metrics.

This analysis for Iraq indicates a decrease in terrorism activity in terms of incidents and casualties in recent years.

```python
# Filter data for Afghanistan, Pakistan, Somalia, and India
afghanistan_data = combined_raw_df[combined_raw_df['Country'] == 'Afghanistan']
pakistan_data = combined_raw_df[combined_raw_df['Country'] == 'Pakistan']
somalia_data = combined_raw_df[combined_raw_df['Country'] == 'Somalia']
india_data = combined_raw_df[combined_raw_df['Country'] == 'India']

# Function to plot trends for a given country
def plot_country_trends(country_data, country_name):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(country_data['Year'], country_data['Incidents'], marker='o', linestyle='-', label='Incidents')
    ax.plot(country_data['Year'], country_data['Fatalities'], marker='s', linestyle='-', label='Fatalities')
    ax.plot(country_data['Year'], country_data['Injuries'], marker='^', linestyle='-', label='Injuries')
    ax.plot(country_data['Year'], country_data['Hostages'], marker='x', linestyle='-', label='Hostages')
    ax.set_title(f'Terrorism Trends in {country_name} (2012-2022)')
    ax.set_xlabel('Year')
    ax.set_ylabel('Count')
    ax.legend()
    plt.xticks(country_data['Year'], rotation=45)
    plt.grid(True)
    plt.show()

# Plotting trends for Afghanistan
plot_country_trends(afghanistan_data, "Afghanistan")
```

The graph above shows the trends in terrorism-related incidents, fatalities, injuries, and hostages in Afghanistan from 2012 to 2022. Here are some observations:

- The number of incidents in Afghanistan has shown significant fluctuations over the years, with notable peaks at certain points.
- Fatalities and injuries also exhibit considerable variation, closely following the pattern of incidents. This suggests a direct correlation between the number of incidents and the number of casualties.
- The data on hostages show spikes in certain years, indicating specific events or situations where hostages were taken in significant numbers.

```python
# Plotting trends for Pakistan
plot_country_trends(pakistan_data, "Pakistan")
```

The trend analysis for Pakistan from 2012 to 2022 reveals:

- A noticeable decline in the number of terrorism-related incidents over the years, indicating a significant improvement in the security situation.
- Fatalities and injuries closely mirror the trend in incidents, with peaks in the earlier years followed by a gradual decrease, suggesting effective counter-terrorism measures may have been implemented.
- The number of hostages has varied year by year, with some years experiencing higher numbers, possibly due to specific kidnapping or hostage-taking incidents

```python
# Plotting trends for Somalia
plot_country_trends(somalia_data, "Somalia")
```

The trend analysis for Somalia from 2012 to 2022 shows:

- A relatively stable number of incidents over the years, with some fluctuations but no clear upward or downward trend.
- Fatalities and injuries follow a somewhat similar pattern to incidents, indicating that the severity of attacks has remained relatively constant.
- The data on hostages show some peaks, suggesting specific years where hostage-taking incidents were more prevalen

```python
# Plotting trends for India
plot_country_trends(india_data, "India")
```



The analysis for India from 2012 to 2022 reveals:

- A relatively stable number of incidents throughout the period, with some annual fluctuations but no clear long-term trend.
- The number of fatalities and injuries also fluctuates year by year, reflecting the variable severity of incidents.
- Hostage situations appear to be less frequent or significant in terms of numbers compared to the other metrics, with minimal fluctuations over the years.

This comprehensive analysis across Iraq, Afghanistan, Pakistan, Somalia, and India highlights varied trends in terrorism's impact over the past decade. Each country exhibits unique patterns in terms of incidents, casualties, and hostage situations, reflecting the complex nature of terrorism and the differing contexts and responses in each country.
